```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
x = ["class","Alcohol" , "Malic acid" , "Ash","Alcalinity of ash", "Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols"
,"Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv("wine.data",names = x )
df
```

```python
import scipy 
from scipy import stats

def skew_function(df,column):
    fig,axes = plt.subplots(1,3,figsize=(15,4))
    sns.boxplot(ax=axes[0],data=df,y=column)
    axes[0].set_title(f"{column} BoxPlot")
    sns.histplot(ax=axes[1],data=df,x=column,kde=True,color='red')
    axes[1].set_title(f"{column} HistPlot")
    scipy.stats.probplot(df[column], sparams=(), dist='norm', fit=True, plot=axes[2], rvalue=False)
    axes[2].set_title(f"{column} ProbabiltityPlot")
    plt.show()
```

```python
for column in df:
    if column!='class':
        skew_function(df,column)  
```
