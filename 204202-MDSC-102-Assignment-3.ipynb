```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import comb
```

# BINOMIAL DISTRIBUTION

```python
def Binomial(n,p):
    x=np.arange(0,n+1,1)
    f = lambda x,n,p: comb(n,x)*p**x * (1-p)**(n-x)      # formula (function)
    y = [f(i,n,p) for i in x]
    plt.bar(x,y)
    plt.title('BINOMIAL DISTRIBUTION')          # keeping the heading
    plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
    plt.ylabel('Y-axis', fontsize = 15)         # label the y axis
    plt.show
Binomial(10,0.5)     #Binomial(n,p)
```

# POISSON DISTRIBUTION

```python
def Poisson(λ):
    e = math.exp(-1*λ)
    f = lambda λ,x  : e * (λ**x)/math.factorial(x)    # formula (function)
    y = [f(λ,i) for i in x]
    plt.bar(x,y)
    plt.title('POISSON DISTRIBUTION')           # keeping the heading
    plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
    plt.ylabel('Y-axis', fontsize = 15)         # label the y axis
    plt.show
Poisson(7)    # poisson(λ)
```

# EXPONENTIAL DISRTIBUTION

```python
def Exponential(λ):
   
    f = lambda λ,x: λ * math.exp((-1*λ)*x)       # formula (function)
    y = [f(λ,i) for i in x ]
    plt.bar(x,y)
    plt.title('EXPONENTIAL DISRTIBUTION')       # keeping the heading
    plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
    plt.ylabel('Y-axis', fontsize = 15)         # label the y axis
    plt.show
Exponential(10)    #Exponential(λ)    
```

## NORMAL DISTRIBUTION

```python
def Normal(m,var):
    f = lambda x,m,var: (1/(math.sqrt(2*(math.pi)*var))) * math.exp((-1/2)*((((x-m))**2)/var))    # formula (function)
    y = [f(i,m,var) for i in x ]
    plt.plot(x,y)
    plt.title('NORMAL DISRTIBUTION')            # keeping the heading
    plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
    plt.ylabel('Y-axis', fontsize = 15)         # label the y axis
    plt.show           
Normal(6,25)   
```

 ## GAMMA DISTIBUTION

```python
def Gamma(a,λ):
    x = np.arange(0,10,1)
    f = lambda x,a,λ: (λ**a)/math.factorial(a-1) * x**(a-1) * math.exp(-λ*x)      # formula (function)
    y = [f(i,a,λ) for i in x]
    plt.plot(x,y)
    plt.title('GAMMA DISRTIBUTION')             # keeping the heading
    plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
    plt.ylabel('Y-axis', fontsize = 15)         # label the y axis
    plt.show 
Gamma(10,30)    
    
    
```
