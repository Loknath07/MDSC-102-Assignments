```python
fig,axis = plt.subplots(4,3,figsize=(15,15))
a,b=0,0
for col in df.columns:
    if col!='class' and col!='Alcohol':
        sns.scatterplot(x='Alcohol',y=col,data=df,hue='class',ax=axis[a,b],palette='deep')
        b+=1
        if b>=3:
            a+=1
            b=0
```
